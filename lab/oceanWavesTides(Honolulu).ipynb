```python
#import essential libraries
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#parameter to use Latex in matplotlib
matplotlib.rcParams.update(matplotlib.rcParamsDefault)
#a plot label should be easy to read
plt.rcParams.update({'font.size': 14})
```

```python
from mpl_toolkits.mplot3d import Axes3D

def linear_vector_field(x, y, z):
    # Define the vector components as functions of x, y, and z
    u = y
    v = 0
    w = 0
    return u, v, w

def diverging_vector_field(x, y, z):
    # Define the vector components as functions of x, y, and z
    norm = x**2 + y**2 + z**2
    dist = np.sqrt(norm)
    return x*dist, y*dist, z*dist

def rotating_vector_field(x, y, z):
    # Define the vector components as functions of x, y, and z
    u = -y
    v = x
    w = 0
    return u, v, w

# Generate data points for the vector field
x, y, z = np.meshgrid(np.linspace(-2, 2, 4), np.linspace(-2, 2, 4), np.linspace(-2, 2, 4))
u, v, w = linear_vector_field(x, y, z)
magnitude = np.sqrt(u**2 + v**2 + w**2)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Color by azimuthal angle
c = y
# Flatten and normalize
c = (c.ravel() - c.min()) / c.ptp()
# Repeat for each body line and two head lines
c = np.concatenate((c, np.repeat(c, 2)))
# Colormap
c = plt.cm.coolwarm(c)

# Plot the vector field
quiver = ax.quiver(x, y, z, u, v, w, length=0.3, color=c, arrow_length_ratio=0.3, ) #normalize = true

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Vector Field')

# Show the plot
plt.show()
```

```python
# Generate data points for the vector field
x, y, z = np.meshgrid(np.linspace(-2, 2, 4), np.linspace(-2, 2, 4), np.linspace(-2, 2, 4))


# Create a 1x3 grid of subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5), subplot_kw={'projection': '3d'})
axs=axs.flatten()

u, v, w = linear_vector_field(x, y, z)
# Color by azimuthal angle
c = y
# Flatten and normalize
c = (c.ravel() - c.min()) / c.ptp()
# Repeat for each body line and two head lines
c = np.concatenate((c, np.repeat(c, 2)))
# Colormap
c = plt.cm.coolwarm(c)
# Plot the vector field with different colors for heads and tails
quiver = axs[0].quiver(x, y, z, u, v, w, length=0.3, color=c , arrow_length_ratio=0.3)

u, v, w = diverging_vector_field(x, y, z)
# Color by azimuthal angle
c = np.sqrt(u**2+v**2+w**2)
# Flatten and normalize
c = (c.ravel() - c.min()) / c.ptp()
# Repeat for each body line and two head lines
c = np.concatenate((c, np.repeat(c, 2)))
# Colormap
c = plt.cm.autumn(c)
# Plot the vector field with different colors for heads and tails
quiver = axs[1].quiver(x, y, z, u, v, w, length=0.3, color=c , arrow_length_ratio=0.3)

u, v, w = rotating_vector_field(x, y, z)
# Color by azimuthal angle
c = z
# Flatten and normalize
c = (c.ravel() - c.min()) / c.ptp()
# Repeat for each body line and two head lines
c = np.concatenate((c, np.repeat(c, 2)))
# Colormap
c = plt.cm.summer(c)
# Plot the vector field with different colors for heads and tails
quiver = axs[2].quiver(x, y, z, u, v, w, length=0.2, color=c , arrow_length_ratio=0.3)

# Plot the vector field in each subplot
for ax in axs:
    # Set labels and title
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    
# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

### Vector potential

Let $\varphi(x,y,z) = x^2 + y^2 + z^2$ be the velocity potential. Compute and draw the corresponding vector field $\bf{v}$. Compute the divergence and the curl of the obtained velocity field and comment the result from the physical point of view.

$\bf{v}= 2 \cdot \begin{pmatrix}x \\ y\\ z\end{pmatrix}$

```python
def vector_field(x, y, z):
    # Define the vector components as functions of x, y, and z
    u = 2*x
    v = 2*y
    w = 2*z
    return u, v, w

# Generate data points for the vector field
x, y, z = np.meshgrid(np.linspace(-2, 2, 4), np.linspace(-2, 2, 4), np.linspace(-2, 2, 4))
u, v, w = vector_field(x, y, z)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Color by azimuthal angle
c = y
# Flatten and normalize
c = (c.ravel() - c.min()) / c.ptp()
# Repeat for each body line and two head lines
c = np.concatenate((c, np.repeat(c, 2)))
# Colormap
c = plt.cm.coolwarm(c)

# Plot the vector field
quiver = ax.quiver(x, y, z, u, v, w, length=0.3, color=c, arrow_length_ratio=0.3, ) #normalize = true

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Vector Field')

# Show the plot
plt.show()
```

This vector field has zero curl yet nonzero divergence. It is similar qualitatively to the second one in the previous exercise.

```python
data=np.loadtxt("honolulu.txt")
signal = data[:,0] - np.mean(data[:,0])
t= data[:,1] -data[0,1]

t= t[signal<2]
signal = signal[signal<2]

# I had to mask some artifacts
exclude_values = [0.26053447408152985, -0.23946552591847015, 0.5105344740815299, 0.010534474081529854]
mask = np.isin(signal, exclude_values, invert=True)
t = t[mask]
signal = signal[mask]

# Perform DFT using FFT
fft_result = np.fft.fft(signal)
freq_axis = np.fft.fftfreq(len(signal))

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(12, 8))

# Plot the original signal
axs[0].scatter(t, signal, s=0.05)
#axs[0].set_xlim(0,2000)
axs[0].set_title('Original Signal')
axs[0].set_xlabel('Time (min)')
axs[0].set_ylabel('Amplitude')

# Plot the frequency spectrum
axs[1].plot(freq_axis[:len(signal)//2], np.abs(fft_result[:len(signal)//2]))
axs[1].set_xscale('log')
axs[1].set_title('Frequency Spectrum')
axs[1].set_xlabel('Frequency (1/min)')
axs[1].set_ylabel('Amplitude')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
#used to build the mask
unique_values, counts = np.unique(signal, return_counts=True)

# Create a list of tuples (value, count) and sort it by count in descending order
sorted_counts = sorted(zip(unique_values, counts), key=lambda x: x[1], reverse=True)

# Display the results
for value, count in sorted_counts:
    print(f"{value}: {count} times")
```

```python
#find the peaks
from scipy.signal import argrelextrema

data = np.abs(fft_result[:len(signal)//2])

# Find the indices of the relative maxima
maxima_indices = argrelextrema(data, np.greater)[0]

# Filter the indices of maxima that are greater than 500
indices_greater_than = maxima_indices[data[maxima_indices] > 800]

# Access the actual values at the maxima indices
maxima_values = data[indices_greater_than]

print("Indices of maxima:", indices_greater_than)
print("Values at maxima:", maxima_values)
```

```python
from scipy.interpolate import interp1d

# create an array with holes just in case:
int_times = np.array(range(43200))
int_signal = np.array([np.nan for _ in range(43200)])
int_signal[t.astype(int)]= signal

# Perform linear interpolation
linear_interpolator = interp1d(t.astype(int), signal, kind='linear', fill_value="extrapolate")
interpolated_t = np.arange(43200)
interpolated_signal = linear_interpolator(interpolated_t)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(t, signal, label='Original Signal', s=3)
plt.scatter(int_times, int_signal, s=0.6, label='Signal with Holes')
plt.scatter(int_times, interpolated_signal, label='Interpolated Signal', s=0.3)
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.legend()
plt.show()
```
